```python
import pandas as pd
```

```python
races = pd.read_csv('data/races.csv')
results = pd.read_csv('data/results.csv')
drivers = pd.read_csv('data/drivers.csv')
constructors = pd.read_csv('data/constructors.csv')
```

```python
final = results.merge(drivers[['driverId', 'driverRef', 'nationality']], how='left', on='driverId')
final = final.merge(constructors[['constructorId', 'constructorRef']], how='left', on='constructorId')
final = final.merge(races[['raceId', 'year', 'name', 'circuitId']], how='left', on='raceId')
```

```python
final
```

```python
final.drop(['position', 'positionText', 'laps', 'time', 'fastestLap', 'milliseconds'], axis=1, inplace=True)
```

```python
final.to_csv('results_dataset.csv')
```
