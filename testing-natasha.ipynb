```python
import pandas as pd
```

```python
races = pd.read_csv('data/races.csv')
results = pd.read_csv('data/results.csv')
drivers = pd.read_csv('data/drivers.csv')
constructors = pd.read_csv('data/constructors.csv')
circuits = pd.read_csv('data/circuits.csv')
```

```python
final = results.merge(drivers[['driverId', 'driverRef', 'forename', 'surname', 'nationality']], how='left', on='driverId')
final = final.merge(constructors[['constructorId', 'name']], how='left', on='constructorId')
final = final.merge(races[['raceId', 'year', 'circuitId']], how='left', on='raceId')
final = final.merge(circuits[['circuitId', 'name']], how='left', on='circuitId', suffixes=('Team', 'Track'))
```

```python
final
```

```python
final.drop(['position', 'positionText', 'laps', 'time', 'fastestLap', 'milliseconds'], axis=1, inplace=True)
```

```python
final
```

```python
final.to_csv('results_dataset.csv')
```

```python
driverStandings = final.copy()
driverStandings
```

```python
driverStandings.drop(['grid','fastestLapTime', 'fastestLapSpeed'], axis=1, inplace=True)
driverStandings
```

```python
driverStandings.drop(driverStandings.loc[driverStandings['positionOrder'] != 1].index, inplace=True)
driverStandings.reset_index(drop=True)
```

```python
firstPlaceCounts = driverStandings.groupby(by=['driverRef', 'positionOrder'])
newCounts = firstPlaceCounts.size()
newCounts.to_csv('first_place_counts.csv')
```

```python
driverStandings.to_csv('first_place_instances.csv')
```
