```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/results.csv', header=0)
# a= np.genfromtxt('data/results.csv', delimiter=',')
```

```python
df
```

```python
fastDict = []
uniqueRaces = df.raceId.unique()
# print(df.iloc[[11141]])
df_no_n = df[df.fastestLapTime.str.replace("\\","") != 'N']
for rID in uniqueRaces:
    rowsByRId = df_no_n[df_no_n.raceId == rID]
    fastestLap = rowsByRId[rowsByRId.fastestLapTime == rowsByRId.fastestLapTime.min()]
    fastDict.append(fastestLap)

result = pd.concat(fastDict)
print(result)
```

```python
result.to_csv('races_fastest_times.csv', index=False)
```
